# Set-up

## Specify directories and file to inspect

```python
f = 'LLAMAS_2026-01-07_01-25-43.5_SCI22'
ex_dir = '/data/llamas/reduced/2026-01-06/extractions/'

fp = {
    "blue":  f"{ex_dir}{f}_extract_RSS_blue.fits",
    "green": f"{ex_dir}{f}_extract_RSS_green.fits",
    "red":   f"{ex_dir}{f}_extract_RSS_red.fits",
}
```

## Import libraries

```python
from astropy.io import fits
import os
```

## Inspect directory

```python
fnames = [f for f in os.listdir(ex_dir) if '.fits' in f]
objnames = []
print('Found files:')
for fname in fnames:
    print(' - ', fname)
    hdul = fits.open(ex_dir+fname)
    print('    OBJECT:', hdul[0].header.get('OBJECT', 'N/A'))
    print('    MJD-OBS:', hdul[0].header.get('MJD-OBS', 'N/A'))
    print('    REXPTIME:', hdul[0].header.get('REXPTIME', 'N/A'))
    objnames.append(hdul[0].header["OBJECT"])
```

## Load reduced 1D spectrum

```python
from utils.io import load_llamas_spectrum
spec_b = load_llamas_spectrum(fp['blue'])
print('Blue band:')
print('wave shape:', spec_b['wave'].shape)
print('flux shape:', spec_b['flux'].shape)
print('fibers:', spec_b['n_fibers'])
```
